```python
import pandas as pd  # importo pandas
from openpyxl import load_workbook  # importo openpyxl para poder abrir y modificar el excel


wb = load_workbook("ejemplo_semana7.xlsx")  # abro el archivo original que subí

hoja = wb.active  # tomo la hoja activa (la primera hoja) para poder modificarla

# Agrego una columna nueca
hoja["C1"] = "Comentario"  # pongo el título de la nueva columna
hoja["C2"] = "Pagado"      # escribo el valor de la fila 2
hoja["C3"] = "No pagado"   # escribo el valor de la fila 3
hoja["C4"] = "No pagado"   # escribo el valor de la fila 4
hoja["C5"] = "Pagado"      # escribo el valor de la fila 5


wb.save("ejemplo_semana7_openpyxl.xlsx")  # guardo el archivo modificado
print("listo")  # pongo listo para saber que ya esta guardado


df = pd.read_excel("ejemplo_semana7_openpyxl.xlsx")  # leo el archivo modificado


df.head()  # muestro las primeras filas del archivo modificado
```
